```python
import pandas as pd
import numpy as np
```

# 讀取逐日spread資料

```python
df_cp = pd.read_csv('/Users/Kang/Desktop/RVol-IVol.csv', index_col = 0)
df_cp = df_cp.set_index(pd.DatetimeIndex(df_cp.index))
df_cp
```

# 將逐日spread轉換成月平均spread

```python
#所有ETF名稱
all_ticker = df_cp.columns.tolist()

#挑出所有年月的最後一天
all_max_day = []
for i in range(df_cp.index.year.min(), df_cp.index.year.max()+1):
    for j in range(1, 13):
        mday = df_cp.loc[(df_cp.index.year == i) & (df_cp.index.month == j)].index.day.max()
        all_max_day.append(str(i) + '-' + ('0'+str(j) if j < 10 else str(j)) + '-' + str(mday))
```

```python
#建立df_m用來儲存每月的平均CVol-PVol
df_m = pd.DataFrame(index = all_max_day, columns = all_ticker)
df_m = df_m.set_index(pd.DatetimeIndex(df_m.index))
```

```python
#計算月平均CVol-PVol並放入df_m中正確的column & index
for i in all_ticker:
    for j in df_m.index:
        df_m[i][j] = df_cp[i].loc[(df_cp.index.year == j.year) & (df_cp.index.month == j.month)].mean()
```

```python
df_m
```

# 對row排序

```python
#assending = True，數值最大者是1，反之最小者是1
#CP用False，RI用True
m_rank = df_m.rank(1, ascending = True)
m_rank
```

```python
#可回傳一個row裡面有幾個NaN
#int(a.isnull().sum(axis=1).values)
#int((a.shape[1] - a.count(axis=1)).values) #效果一樣
#m_rank.isnull().sum(axis=1).values.max() #最多有兩個NaN而已
```

# 將排序轉換為權重

```python
#主觀權重配置
e11 = [0.4, 0.3, 0.2, 0.1, 0, 0, 0, -0.1, -0.2, -0.3, -0.4]
e10 = [0.4, 0.3, 0.2, 0.1, 0, 0, -0.1, -0.2, -0.3, -0.4] #另一個NaN也是0
e9 = [0.4, 0.3, 0.2, 0.1, 0, -0.1, -0.2, -0.3, -0.4] #另兩個NaN都是0
#e9 = [(3/6), (2/6), (1/6), 0, 0, 0, (-1/6), (-2/6), (-3/6)] #另兩個NaN都是0
```

```python
#每輪到一行，都先算出那行有幾個NaN
for day in m_rank.index:
    tem = m_rank.loc[(m_rank.index.year == day.year) & (m_rank.index.month == day.month)]
    nan_count = int(tem.isnull().sum(axis=1).values)
    
    if nan_count == 0:
        #根據事先設定的權重，用
        for ticker in all_ticker:
            m_rank[ticker][day] = e11[int(m_rank[ticker][day]) - 1]
            
    elif nan_count == 1:
        for ticker in all_ticker:
            #如果是NaN就直接填入0
            if np.isnan(m_rank[ticker][day]):
                m_rank[ticker][day] = 0
            else:
                m_rank[ticker][day] = e10[int(m_rank[ticker][day]) - 1]
            
    elif nan_count == 2:
        for ticker in all_ticker:
            if np.isnan(m_rank[ticker][day]):
                m_rank[ticker][day] = 0
            else:
                m_rank[ticker][day] = e9[int(m_rank[ticker][day]) - 1]
```
