```python
import pandas as pd
from datetime import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

# 寫入檔案、設定權重

```python
df = pd.read_csv('/Users/Kang/Desktop/AGG、EMB、EWT、IVV、JNK、PFF、TLT、VGK、VNQ、VPL、VWO.csv', )
df.set_index("Date", inplace = True)
df
```

```python
w0 = {'AGG' : 0.1, 'EMB':0.05, 'EWT':0.1, 'IVV':0.2, 'JNK':0.05, 'PFF':0.05, 'TLT':0.05, 'VGK':0.1, 'VNQ':0.1, 'VPL':0.1, 'VWO':0.1}
```

# 挑出所有rebalance日，並新增df_w以及NAV欄

```python
df = df.set_index(pd.DatetimeIndex(df.index))

#可得到每年每月最後一個交易日
def get_max_date_each_month(df):
    tempTradeDays = df.index
    dateRange = []  
    tempYear = None
    #先將交易日以所有資料的年份分群
    dictYears = tempTradeDays.groupby(tempTradeDays.year)
    for yr in dictYears.keys():
        #再將每一年依序以月份分群
        tempYear = pd.DatetimeIndex(dictYears[yr]).groupby(pd.DatetimeIndex(dictYears[yr]).month)
        #逐月將當月最大那日放到dateRange裡
        for m in tempYear.keys():
            dateRange.append(max(tempYear[m]))
    dateRange = pd.DatetimeIndex(dateRange).sort_values()
    return dateRange
```

```python
largest_date = get_max_date_each_month(df) #每年每月最後一個交易日
df_w = pd.DataFrame(index = largest_date, columns = df.columns.tolist()) #建一個空的df_w用來儲存每個rebalance的新權重
df['NAV'] = np.nan #建一個新欄NAV，做了以上步驟後才能新增此欄
```

# 先算逐日NAV，再算rebalance日的新權重（最開始日是算權重再算NAV）

```python
price_to_balance = 10000 #begining invest
tem_count = 0

#用以計算到Rebalance當天的NAV
def caculate_nav(i, tem_count): #計算到第幾個rebalance day
    while tem_count < len(df): #當tem_count等於len(df)時不能跑，而且事實上也已經跑完了
        #對於比第i個權重日期還要早或相同的所有日期
        if largest_date[i] >= df.index[tem_count]:
            rebalance_day = dt.strftime(largest_date[i - 1].date(), '%Y-%m-%d') #是使用上一個rebalance日的權重
            df['NAV'].iloc[tem_count] = sum(df[j].iloc[tem_count] * df_w[j].loc[rebalance_day] for j in df_w.columns.tolist()) #計算每天的NAV
            tem_count += 1
        else:
            break  
    return tem_count

def rebalance_weight(j): #計算到第幾個rebalance day
    t = dt.strftime(largest_date[j].date(), '%Y-%m-%d') 
    for etf in df_w.columns.tolist():
        df_w[etf][t] = price_to_balance * w0[etf] / df[etf][t] #計算出Rebalance日所有etf的新權重
```

```python
#先跑最開始權重計算以及算NAV
rebalance_weight(0)
rebalance_day = dt.strftime(largest_date[0].date(), '%Y-%m-%d')
df['NAV'].iloc[tem_count] = sum(df[j].iloc[tem_count] * df_w[j].loc[rebalance_day] for j in df_w.columns.tolist())
tem_count += 1

#之後在每一個rebalance日之前，都先算出所有NAV，然後用rebalance當天的etf價格，重新分配權重
for day in range(1, len(df_w)): #第幾個rebalance day，第0個已經在上面先跑過了，因為流程跟這邊不一樣
    tem_count = caculate_nav(day, tem_count) #將自己循環代入函式，可不斷得到新的斷位點，也就是下次rebalance的前一天
    price_to_balance = df['NAV'][tem_count - 1] #因為tem_count加一，需減一才是Rebalance當天的NAV
    rebalance_weight(day)
```
